```python
#@title Gaussian Elemenation

# Importing NumPy Library
import numpy as np
import sys


def gaussian_elimination(matrices):
  n=len(matrices)
# Making numpy array of n size and initializing 
# to zero for storing solution vector
  output = np.zeros(n)

  # Applying Gauss Elimination
  for i in range(n):
      if matrices[i][i] == 0.0:
          sys.exit('Divide by zero detected!')
          
      for j in range(i+1, n):
          ratio = matrices[j][i]/matrices[i][i]
          
          for k in range(n+1):
              matrices[j][k] = matrices[j][k] - ratio * matrices[i][k]
  print("forward sub is ",matrices)

  # Back Substitution
  output[n-1] = matrices[n-1][n]/matrices[n-1][n-1]

  for i in range(n-2,-1,-1):
      output[i] = matrices[i][n]
      
      for j in range(i+1,n):
          output[i] = output[i] - matrices[i][j]*output[j]
      
      output[i] = output[i]/matrices[i][i]
  # Displaying solution
  print('\nRequired solution is: ')
  if len(matrices)==1:
    names=['x']
  if len(matrices)==2:
    names=['x','y']
  if len(matrices)==3:
    names=['x','y','z']
  if len(matrices)==4:
    names=['x','y','z','w']
  if len(matrices)==5:
    names=['x','y','z','w','m']  
  if len(matrices)==6:
    names=['x','y','z','w','m','n']  
  if len(matrices)==7:
    names=['x','y','z','w','m','n','k']  
  if len(matrices)==8:
    names=['x','y','z','w','m','n','k','a']
  if len(matrices)==9:
    names=['x','y','z','w','m','n','k','a','b']
  if len(matrices)==10:
    names=['x','y','z','w','m','n','k','a','b','c']
  if len(matrices)==11:
    names=['x','y','z','w','m','n','k','a','b','c','d']
 

  for i in range(n):

      print('%c = %0.2f' %(names[i],output[i]), end = '\t')
```
